# What is meant ouliers?

Outliers are extreme values in a dataset that significantly differ from most other points and can distort statistical results.

✅ Example

If most students score between 60–90, but one student scores 5 or 100, those could be considered outliers.

✅ Why outliers matter?

They can reduce model accuracy

They can skew averages and trends

They may indicate errors, rare events, or important insights

✅ How to detect them?

Z-Score

IQR method (Q1 – 1.5 IQR, Q3 + 1.5 IQR)

Boxplot visualizations
