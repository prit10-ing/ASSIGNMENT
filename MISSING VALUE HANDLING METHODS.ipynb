# MISSING VALUE HANDLING METHODS

In Pandas, handling missing values is a very common task. Missing values are usually represented by NaN (Not a Number).

#### üîç 1. Detection of Missing Values
‚úÖ Functions:

df.isnull() ‚Üí Returns a DataFrame of True/False for missing values.

df.isna() ‚Üí Same as isnull().

df.notnull() or df.notna() ‚Üí Returns True where values are not missing.

#### 2. Removing Missing Values
‚úÖ Function: dropna()

Parameter Description


axis=0	    Removes rows with missing values

axis=1	    Removes columns with missing values

how='any'	Removes if any value is missing (default)

how='all'	Removes if all values are missing

subset=     ['col1','col2']	Only checks specific columns

inplace=    True Updates the DataFrame directly

#### üß© 3. Filling Missing Values
‚úÖ Function: fillna(value, method, inplace)

Used to replace missing values with a specific value, statistic, or nearby value.

(a) Fill with a Specific Value
df['age'].fillna(0, inplace=True)








(b) Fill with Mean / Median / Mode

Statistic	Usage	Best for

Mean	df['col'].fillna(df['col'].mean())	Normally distributed data

Median	df['col'].fillna(df['col'].median())	Skewed data


Mode	df['col'].fillna(df['col'].mode()[0])











(c) Forward Fill (ffill)

df.fillna(method='ffill', inplace=True)

üëâ Fills missing value with previous value (useful for time-series).

(d) Backward Fill (bfill)

df.fillna(method='bfill', inplace=True)
üëâ Fills missing value with next value.


#### 4. Random Imputation

Randomly replace missing values with random samples from the same column (useful for large datasets).



import numpy as np

df['age'].fillna(np.random.choice(df['age'].dropna()), inplace=True)

üëâ Randomly fills missing ‚Äúage‚Äù values with existing ones.

#### 5. Predictive Imputation (Machine Learning Based)

Used when missing values are many and you want more accurate imputation.

‚úÖ Example using KNN:


from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=3)
df[['col1', 'col2']] = imputer.fit_transform(df[['col1', 'col2']])



üëâ Predicts missing values using nearest neighbors.
